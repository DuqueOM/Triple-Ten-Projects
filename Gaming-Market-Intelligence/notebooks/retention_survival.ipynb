# Retention & Survival Analysis (Kaplanâ€“Meier)
Compute survival curve and simple LTV estimate.

```python
import numpy as np
import pandas as pd
from lifelines import KaplanMeierFitter

# Synthetic retention cohort
rng = np.random.default_rng(0)
n = 1000
durations = rng.exponential(scale=6, size=n)  # months
events = rng.binomial(1, 0.7, size=n)
km = KaplanMeierFitter()
km.fit(durations, event_observed=events, label='Retention')
km.survival_function_.head()
```

```python
# Simple LTV: sum over expected revenue per month weighted by survival
monthly_rev = 5.0  # $5 per active user
sf = km.survival_function_['Retention']
ltv = float(np.sum(sf.values * monthly_rev))
ltv
```
