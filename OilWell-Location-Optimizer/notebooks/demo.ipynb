Quick demo: load/exported models, predict on a sample, and inspect saved metrics.

```python
from pathlib import Path
import json
import pandas as pd

from app.example_load import load_models

print('Imports ready')
```

```python
models = load_models()
if not models:
    print('Models not found. Run training first: !python main.py --mode train --config configs/default.yaml')
else:
    region = sorted(models.keys())[0]
    model = models[region]
    sample = pd.DataFrame([{'f0': 1.0, 'f1': -2.0, 'f2': 3.0}])
    pred = float(model.predict(sample)[0])
    print(json.dumps({'region': region, 'prediction': pred}, indent=2))
```

```python
metrics_path = Path('artifacts/metrics.json')
if metrics_path.exists():
    print('Metrics:')
    print(metrics_path.read_text())
else:
    print('No metrics found. Generate with: !python main.py --mode train --config configs/default.yaml')
```

```python
risk_path = Path('artifacts/risk_results.json')
if risk_path.exists():
    print('Risk results:')
    print(risk_path.read_text())
else:
    print('No risk results found. Generate with: !python main.py --mode eval --config configs/default.yaml')
```
